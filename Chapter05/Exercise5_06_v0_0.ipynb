```python
OPEN_DATA_URL = './data/canada-open-data/inventory.csv'
COLUMNS = ['title_en', 'description_en','date_released']
catalog = pd.read_csv(OPEN_DATA_URL, usecols=COLUMNS)
catalog = catalog.dropna(subset=['description_en'])
catalog
```

```python
catalog.shape
```

```python
import re

HANDLE = '@\w+'
LINK = 'https?://t\.co/\w+'
SPECIAL_CHARS = '&lt;|&lt;|&amp;|#'
def clean(text):
    #text = re.sub(HANDLE, ' ', text)
    text = re.sub(LINK, ' ', text)
    text = re.sub(SPECIAL_CHARS, ' ', text)
    return text

catalog['description_en'] = catalog.description_en.apply(clean)
catalog.head(10)
```

```python
catalog['description_en'].sample(frac=0.25,replace=False,random_state=0).to_csv('./data/canada-open-data/catalog.txt',encoding='utf-8')
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
file='./data/canada-open-data/catalog.txt'
f=open(file,'r',encoding='utf-8')
text=f.read()
f.close()
```

```python
text
```

```python
doc=nlp(text)
```

```python
pos_list=['NOUN'] #We can experiment with other or a combinations of parts of speech ['NOUN','ADJ','VERB','ADV'] #['NOUN','ADJ']
```

```python
preproc_text=[]
preproc_sent=[]

for token in doc:
    if token.text!='\n':
        if not(token.is_stop) and not(token.is_punct) and token.pos_ in pos_list:
            preproc_sent.append(token.lemma_)
    else:
        preproc_text.append(preproc_sent)
        preproc_sent=[]

preproc_text.append(preproc_sent) #last sentence

print(preproc_text)
```

```python
len(preproc_text)
```

## HDP

```python
import tomotopy as tp
```

```python
mdl = tp.HDPModel(alpha=0.1,eta=0.20,seed=0)

for line in preproc_text:
    mdl.add_doc(line)

for i in range(0, 110, 10):
    mdl.train(i)
    print('Iteration: {}\tLog-likelihood: {}'.format(i, mdl.ll_per_word))

for k in range(mdl.k):
    print('Top 10 words of topic #{}'.format(k))
    print(mdl.get_topic_words(k, top_n=7))
```

```python
bag_of_words=[word for sent in preproc_text for word in sent]
```

```python
doc_inst = mdl.make_doc(bag_of_words)
mdl.infer(doc_inst)[0]
```

```python
np.argsort(np.array(mdl.infer(doc_inst)[0]))[::-1]
```
