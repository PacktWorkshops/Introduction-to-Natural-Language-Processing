## Summarize complaints in Consumer Financial Protection Bureau dataset using fine-tuned pre-trained BERT Extractive Summarizer (bert-extractive-summarizer) and Text Rank (gensim) 
# https://www.consumerfinance.gov/data-research/consumer-complaints/

```python
import os
import csv
import pandas as pd
```

```python
from gensim.summarization import summarize
```

```python
from summarizer import Summarizer
```

```python
model = Summarizer(hidden=-2,random_state=12345)
```

```python
complaints_pathname='./data/consumercomplaints/Consumer_Complaints.csv'
df_all_complaints=pd.read_csv(complaints_pathname)
```

```python
print('The total number of complaints in the dataset is ', df_all_complaints.shape[0])
 #df_all_complaints.columns
```

```python
df_all_narr=df_all_complaints.dropna(subset=['Consumer complaint narrative'])
df_all_narr=df_all_narr[['Product','Sub-product','Issue','Sub-issue','Consumer complaint narrative']]
print('The total number of complaints with narrative in the dataset is ', df_all_narr.shape[0])
```

```python
df_part_narr=df_all_narr.loc[[242830,1086741,536367,957355,975181,483530,950006,865088,681842,536367,132345,285894]]
```

```python
df_part_narr
```

```python
df_part_narr['BERT Extractive Summary']=df_part_narr['Consumer complaint narrative'].apply(lambda x: model(x,min_length=0,ratio=0.20))
```

```python
def try_summarize(x,ratio):
    try:
        return(summarize(x,ratio=ratio))
    except:
        return('')
```

```python
df_part_narr['TextRank Summary']=df_part_narr['Consumer complaint narrative'].apply(lambda x: try_summarize(x,ratio=0.20))
```
